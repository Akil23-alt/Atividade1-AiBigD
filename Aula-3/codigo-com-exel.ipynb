```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('monitoramento_tempo.csv',sep=',')
```

```python
df.head(15)
```

```python
df.info()
```

```python
import datetime 
df['data'] = pd.to_datetime(df['data'])
```

```python
df.info()
```

```python
df.describe()
```

```python
plt.figure(figsize=(15,8))
plt.plot(df['data'],df['temperatura'])
plt.xlabel('Data e hora',fontsize=18)
plt.ylabel('Temperatura',fontsize=18)
```

```python
fig = plt.figure(figsize=(10,8))
eixo = fig.add_axes([0,0,1,1])
eixo.plot(df['data'],df['temperatura']/10,color='red',lw=0.2,linestyle='--')
eixo.legend(['temperatura'],loc='lower right' ,fontsize=15)
eixo.set_xlebel('Ano',fontsize=20)
eixo.set_ylebel('Temperatura',fontsize=20)
```

```python
fig = plt.figure(figsize=(15,8))
plt.rcParams['font.size']='20'
eixo = fig.add_axes([0,0,1,1])
eixo.plot(df['data'],df['temperatura'],color='b',lw=0.5,linestyle='--')
eixo.set_xlim(datetime.datetime(2015,1,1),datetime.datetime(2015,12,31))
eixo.set_title('Temperatura - Ano 2015',fontsize=25)
eixo.legend(['temperatura'],fontsize=20)
eixo.set_xlabel('Ano',fontsize=20)
plt.xticks(rotation=45,fontsize=15,ha='right')
plt.show 

```
