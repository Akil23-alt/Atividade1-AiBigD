```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
Dados_x = np.linspace(0,10,10)
Dados_y = Dados_x * Dados_x
```

```python
plt.figure(figsize=(10,8),facecolor='gray')
Ax = plt.axes()
Ax.set_facecolor('white')
plt.plot(Dados_x,Dados_y,color='red',marker='o',linestyle='dashed',markeredgecolor='black')
plt.xlabel('Eixo x',fontsize=16)
plt.ylabel('Eixo y',fontsize=16)
plt.title('Grafico simples com Matplotlib',fontsize=16)
```

```python
plt.rcParams.update({'font.size':15})
plt.figure(figsize=(12,12))
linhas = 2
colunas = 2
plt.subplot(linhas,colunas,1)
plt.plot(Dados_x,'r--',label='Dados_x')
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.title('Grafico 1')
plt.legend()

plt.subplot(linhas,colunas,2)
plt.plot(Dados_x,Dados_y,'b--',label='Dados_x \n e y')
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.title('Grafico 2')
plt.legend()

plt.subplot(linhas,colunas,3)
plt.plot(Dados_x*Dados_x,'g--',label='Dados_x^2')
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.title('Grafico 3')
plt.legend()

plt.subplot(linhas,colunas,4)
plt.plot(Dados_y,'c--',label='Dados_y')
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.title('Grafico 4')
plt.legend()
plt.savefig('Graficos aula 03.png')
plt.show()
```

```python
Figura = plt.figure()
Grafico = Figura.add_axes([0,0,0.7,0.7])
Rotulos =['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J']
Grafico.bar(Rotulos,Dados_x,color='red')
Grafico.set_xlabel('Rotulos')
Grafico.set_ylabel('Quantidade')
```
